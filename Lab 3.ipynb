```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

# Preview data
print(df.head())
```

```python
print("Missing values:",df.notnull().sum())
```

```python
num_cols = ['math score', 'reading score', 'writing score']
```

```python
for col in num_cols:
    print(f"\n{col}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])

print(df['writing score'].mean())
```

```python
for col in num_cols:
    print(f"\n{col}")
    print("Min:", df[col].min())
    print("Max:", df[col].max())
    print("Sum:", df[col].sum())
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())
```

```python
for col in num_cols:
    print(f"\n{col}")
    print("Q1:",df[col].quantile(0.25))
    print("Q2 (median):",df[col].quantile(0.5))
    print("Q3:",df[col].quantile(0.75))
```

```python
correlation = df[num_cols].corr()
print("Correlation Matrix:\n", correlation)

covariance = df[num_cols].cov()
print("\nCovariance Matrix:\n", covariance)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['math score'])
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['math score'])
plt.title("Boxplot of Math Scores")
plt.show()
```
